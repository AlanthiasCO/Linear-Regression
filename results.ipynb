```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
report_df = pd.read_csv('data_and_metrics//model_performance_report_all_models.csv')

report_df.head()
```

```python
# Plotar comparação de MSE
plt.figure(figsize=(8, 4))
mse_data = report_df[report_df['Metric'] == 'Mean Squared Error (MSE)']
plt.bar(mse_data['Model'], mse_data['Value'], color='blue')
plt.title('Comparação de Mean Squared Error (MSE)')
plt.xlabel('Modelos')
plt.ylabel('MSE')
plt.xticks(rotation=45)
plt.savefig(r'C:\Users\Alan\projects\home_prices(linear_reg)\images\mse_comparison.png', bbox_inches='tight')
plt.show()

```

```python
# Plotar comparação de R²
plt.figure(figsize=(8, 4))
r2_data = report_df[report_df['Metric'] == 'R-squared (R²)']
plt.bar(r2_data['Model'], r2_data['Value'], color='green')
plt.title('Comparação de R-squared (R²)')
plt.xlabel('Modelos')
plt.ylabel('R²')
plt.xticks(rotation=45)
plt.savefig(r'C:\Users\Alan\projects\home_prices(linear_reg)\images\r2_comparison.png', bbox_inches='tight')
plt.show()
```

```python
# Plotar comparação de MAPE
plt.figure(figsize=(8, 4))
mape_data = report_df[report_df['Metric'] == 'Mean Absolute Percentage Error (MAPE)']
plt.bar(mape_data['Model'], mape_data['Value'], color='red')
plt.title('Comparação de Mean Absolute Percentage Error (MAPE)')
plt.xlabel('Modelos')
plt.ylabel('MAPE (%)')
plt.xticks(rotation=45)
plt.savefig(r'C:\Users\Alan\projects\home_prices(linear_reg)\images\mape_comparison.png', bbox_inches='tight')
plt.show()
```

Teste de subplot:

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plotar comparação de MSE
mse_data = report_df[report_df['Metric'] == 'Mean Squared Error (MSE)']
axes[0].bar(mse_data['Model'], mse_data['Value'], color='blue')
axes[0].set_title('Comparação de Mean Squared Error (MSE)')
axes[0].set_xlabel('Modelos')
axes[0].set_ylabel('MSE')
axes[0].tick_params(axis='x', rotation=45)

# Plotar comparação de R²
r2_data = report_df[report_df['Metric'] == 'R-squared (R²)']
axes[2].bar(r2_data['Model'], r2_data['Value'], color='green')
axes[2].set_title('Comparação de R-squared (R²)')
axes[2].set_xlabel('Modelos')
axes[2].set_ylabel('R²')
axes[2].tick_params(axis='x', rotation=45)
axes[2].set_ylim(0, max(r2_data['Value']) * 1.1) # feito para garantir que todos os 
                                                 # valores caibam no gráfico

# Plotar comparação de MAPE
mape_data = report_df[report_df['Metric'] == 'Mean Absolute Percentage Error (MAPE)']
axes[1].bar(mape_data['Model'], mape_data['Value'], color='red')
axes[1].set_title('Comparação de Mean Absolute Percentage Error (MAPE)')
axes[1].set_xlabel('Modelos')
axes[1].set_ylabel('MAPE (%)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig(r'C:\Users\Alan\projects\home_prices(linear_reg)\images\overview.png', bbox_inches='tight')
plt.show()
```
