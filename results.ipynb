```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
report_df = pd.read_csv('model_performance_report_all_models.csv')

report_df.head()
```

```python
# Plotar comparação de MSE
plt.figure(figsize=(8, 4))
mse_data = report_df[report_df['Metric'] == 'Mean Squared Error (MSE)']
plt.bar(mse_data['Model'], mse_data['Value'], color='blue')
plt.title('Comparação de Mean Squared Error (MSE)')
plt.xlabel('Modelos')
plt.ylabel('MSE')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotar comparação de R²
plt.figure(figsize=(8, 4))
r2_data = report_df[report_df['Metric'] == 'R-squared (R²)']
plt.bar(r2_data['Model'], r2_data['Value'], color='green')
plt.title('Comparação de R-squared (R²)')
plt.xlabel('Modelos')
plt.ylabel('R²')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotar comparação de MAPE
plt.figure(figsize=(8, 4))
mape_data = report_df[report_df['Metric'] == 'Mean Absolute Percentage Error (MAPE)']
plt.bar(mape_data['Model'], mape_data['Value'], color='red')
plt.title('Comparação de Mean Absolute Percentage Error (MAPE)')
plt.xlabel('Modelos')
plt.ylabel('MAPE (%)')
plt.xticks(rotation=45)
plt.show()
```
