```python
import pandas as pd
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
with open('data.pkl', 'rb') as f:
    data = pkl.load(f)
```

```python
# set again:
house_sizes = data ['HouseSize'].values
house_prices = data['HousePrice'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(house_sizes, house_prices, test_size=0.2, random_state=42)
```

```python
# Reshape the data for NumPy
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Predict prices for the test set
predictions = model.predict(x_test)

# Visualize the predictions
plt.scatter(x_test, y_test, marker='o', color='blue', label='Preços Reais')
plt.plot(x_test, predictions, color='red', linewidth=2, label='Previsões de Preços')
plt.title('Previsão de Preços de Propriedades com Regressão Linear')
plt.xlabel('Tamanho da Casa(sq.ft)')
plt.ylabel('Preço da Casa (milhões $)') # em dolares
plt.legend()
plt.show()
```

O gráfico gerado mostra a relação entre o tamanho da casa (em pés quadrados) e o preço da casa (em milhões de dólares).

Os pontos azuis representam os preços reais das casas no conjunto de teste. A linha vermelha representa os preços previstos pelo modelo de regressão linear.

Se a linha vermelha (previsões) estiver próxima dos pontos azuis (valores reais), isso indica que o modelo está fazendo boas previsões.

Se houver uma grande discrepância entre a linha vermelha e os pontos azuis, isso sugere que o modelo pode não estar capturando bem a relação entre o tamanho da casa e o preço.
